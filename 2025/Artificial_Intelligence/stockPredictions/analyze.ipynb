```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_data():
    raw_data = dict()
    with open("bigger_data_set.txt", "r") as f:
        for line in f:
            line = line.split(" ")
            raw_data[line[0]] = list(map(float, line[1:]))
    return pd.DataFrame.from_dict(raw_data)
```

```python
df = read_data()
df.head()
```

```python
df.index
```

```python
fig, ax = plt.subplots(figsize=(15, 9))
for i in df.columns:
    ax.plot(df.index, df.loc[:, i], label=i)
leg = ax.legend(loc = "upper left")
```

```python
fig = plt.figure(figsize=(15,9))
idx = 1
for i in df.columns:
    ax = fig.add_subplot(3, 3, idx)
    ax.plot(df.index, df.loc[:, i], label=i)
    leg = ax.legend(loc = "upper left")
    idx += 1
```
